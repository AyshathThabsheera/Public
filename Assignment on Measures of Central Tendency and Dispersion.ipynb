# 2. What do you think about the students' parental level of education?

```python
plt.title('students parental level of education')
sns.countplot(x='parental level of education', data=data)
plt.xticks(rotation=90)
```

```python
plt.title('Parentel level of Education',fontsize = 20)
data['parental level of education'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

Majority of student's parents are having high school level of education or educated from some colleges or having associate degree.A few student's parents is having master's degree

# 3. Who scores the most on average for math, reading and writing based on
#  Gender
#  Test preparation course

```python
#based on gender
```

```python
data.groupby('gender')[['math score','reading score','writing score']].mean()
```

Male students scores more in maths whereas female students scores in both reading and writing.

```python
#based on test preparation course
```

```python
data.groupby('test preparation course')[['math score','reading score','writing score']].mean()
```
