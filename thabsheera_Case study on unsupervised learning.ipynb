```python
num_cols = ['Alcohol', 'Malic_Acid', 'Ash', 'Ash_Alcanity', 'Magnesium',
       'Total_Phenols', 'Flavanoids', 'Nonflavanoid_Phenols',
       'Proanthocyanins', 'Color_Intensity', 'Hue', 'OD280', 'Proline']

for i in num_cols:
    plt.figure(figsize=(5,2.92))
    plt.boxplot(data[i])
    plt.title(i)
```

```python
Q1 = np.percentile(data['Malic_Acid'],25,interpolation='midpoint')
Q2 = np.percentile(data['Malic_Acid'],50,interpolation='midpoint')
Q3 = np.percentile(data['Malic_Acid'],75,interpolation='midpoint')

IQR = Q3 - Q1

up_lim = Q3 + 1.5*IQR
low_lim = Q1 - 1.5*IQR

data['Malic_Acid'] = np.where(data['Malic_Acid'] > up_lim, up_lim, np.where(data['Malic_Acid'] < low_lim, low_lim, data['Malic_Acid']))

plt.figure(figsize=(5, 4))
plt.boxplot(data['Malic_Acid'])
plt.title('Boxplot of Malic_Acid after Handling Outliers')
plt.show()
```

```python
#scaling
```

```python
from sklearn.preprocessing import StandardScaler

scaler =StandardScaler()
 
features =scaler.fit(data)
features =features.transform(data)
 
# Convert to pandas Dataframe
data1 =pd.DataFrame(features,columns=data.columns)
# Print the scaled data
data1.head()
```

```python
#kmeans clustering
```

```python
X=data1.values
```

```python
from sklearn.cluster import KMeans

wcss = {}
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss[i] = kmeans.inertia_
     
plt.plot(wcss.keys(), wcss.values(), 'gs-')
plt.xlabel("Values of 'k'")
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
labels = kmeans.labels_


# Add a new column for customer types based on cluster labels
data['Cluster'] = labels
```

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Wine Cluster')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg1 = silhouette_score(X, y_kmeans)
print(silhouette_avg1)
```

```python
#PCA
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
 
reduced_X=pd.DataFrame(data=pca.fit_transform(data1),columns=['PCA1','PCA2'])
 
#Reduced Features
reduced_X.head()
```

```python
centers=pca.transform(kmeans.cluster_centers_)
```

```python
plt.figure(figsize=(7,5))
 
# Scatter plot
plt.scatter(reduced_X['PCA1'],reduced_X['PCA2'],c=kmeans.labels_)
plt.scatter(centers[:,0],centers[:,1],marker='x',s=100,c='red')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.title('Wine Cluster')
plt.tight_layout()
```

```python
from sklearn.metrics import silhouette_score,calinski_harabasz_score, davies_bouldin_score

silhouette_avg2 = silhouette_score(reduced_X, y_kmeans)
print(silhouette_avg2)
```

```python
#agglomerative heirarchical clustering
```

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Wines')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')
y = hc.fit_predict(X)
```

```python
plt.scatter(X[y==0,0], X[y==0,1], c='red', s=50)
plt.scatter(X[y==1,0], X[y==1,1], c='blue', s=50)
plt.scatter(X[y==2,0], X[y==2,1], c='green', s=50)
plt.title('heirarchical Clustering')
plt.xlabel('Alcohol')
plt.ylabel('Malic_Acid')
plt.show()
```

```python
silhouette_avg3 = silhouette_score(X, y)
print(silhouette_avg3)
```

```python
#DBSCAN
```

```python
from sklearn.neighbors import NearestNeighbors
# we use sklearn to find the nearest neighbors
neigh = NearestNeighbors(n_neighbors=10)
nbrs = neigh.fit(data1)
distances, indices = nbrs.kneighbors(data1)
for k in range(1, 10):
    distances_k = np.sort(distances[:,k], axis=0)
    plt.plot(distances_k, label='minPts=%i' % (k+1))
sns.set_style('darkgrid')
plt.ylabel('epsilon')
plt.xlabel('Number of core points')
plt.axhline(2.9)
#plt.axhline(2.7)
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=3 ,min_samples=2)
dbscan.fit(data1)
clusters = dbscan.labels_
len(set(clusters))
```

```python
 clusters
```

```python
# Data points count per label
unique, counts = np.unique(clusters, return_counts=True)
np.asarray((unique, counts)).T
```

```python
s = silhouette_score(data1,clusters)
c = calinski_harabasz_score(data1, clusters)
d = davies_bouldin_score(data1, clusters) 
s,c,d
```

```python
from sklearn import decomposition

# Transform data to visualize data in 2d
sns.set_style('dark')
pca = decomposition.PCA(n_components=3)
wine_pca = pca.fit_transform(data1)
plt.figure(figsize=(9,7))
sns.scatterplot(hue=dbscan.labels_, x=wine_pca[:,0], y=wine_pca[:,1],palette="deep")
plt.title("DBSCAN with three clusters",fontsize=15,fontweight="bold")
plt.show()
```
